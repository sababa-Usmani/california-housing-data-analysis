```python
#importing necessary libraries
import pandas as pd #for data analysis,cleaning,exploration and manipulation
import matplotlib.pyplot as plt #for data visualization and presentation

#Load the Dataset
data=pd.read_csv("/content/housing+(1) (1).csv") #Read the scv into data

```

1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python
#Calculte the average median income of data
avg_median_income=data['median_income'].mean() #mean(): to calculate average median_income
print(f'Average median income:{avg_median_income}')

#Distribution of median_income(using histogram)
plt.figure(figsize=(8, 4)) #set dimension of the figure
plt.hist(data['median_income'],color='lightblue',edgecolor='k') #plot histogram
plt.xlabel('Median Income') #label the x-axis
plt.ylabel('frequency') #label the y-axis
plt.title('Histogram of Median Income') #title of histogram

plt.show()#display figure
```

Explanation : The above histogram is showing the distribution of the median Income of across the data set.
The plot represents a rightly skewed data or positive skewed data where mean is greater than the median. Also, majority of the data is falling on the right side of the median.

6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
'''A user-defined function calculate_median that takes a dataset and a column name as input and
returns the median value of that column.
This function can be reused to calculate medians wherever needed.'''

#Define a function to calculate median
def calculate_median(data,column):
  return data[column].median()

#Examples of Usage for different columns
med_housing_median_age=calculate_median(data,'housing_median_age')
med_median_income=calculate_median(data,'median_income')

print(f'Median of housing median age:{med_housing_median_age}')
print(f'Median of median income:{med_median_income}')
```

10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
#create new column 'total_bedroom_size'

# Define a function to categorize bedroom sizes
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 10 < total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

data['total_bedroom_size'] = data['total_bedrooms'].apply(categorize_bedroom_size) #create new column
data #displaythe data set with new column
```
